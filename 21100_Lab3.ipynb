```python
import librosa
import librosa.display
import IPython.display as ipd

# Step 1: Load the audio file
filename = 'Statement.wav'  
signal, sr = librosa.load(filename, sr=None)

# Listen to the original audio
print("Original Audio:")
ipd.display(ipd.Audio(filename))

# Step 2: Trim the silence
trimmed_signal, index = librosa.effects.trim(signal, top_db=20)  

# Step 3: Listen to the new signal
print("Trimmed Audio:")
ipd.display(ipd.Audio(trimmed_signal, rate=sr))
```

```python
import librosa
import soundfile as sf
import numpy as np 

audio_file = 'Statement.wav'
y, sr = librosa.load(audio_file, sr=None)

top_db_values = [30, 20, 10] 

for top_db in top_db_values:
    intervals = librosa.effects.split(y, top_db=top_db)

    split_segments = []

    for interval in intervals:
        split_segments.append(y[interval[0]:interval[1]])

    split_audio = np.concatenate(split_segments)

    print(f"Split Audio (top_db={top_db}):")
    ipd.display(ipd.Audio(split_audio, rate=sr))
```
