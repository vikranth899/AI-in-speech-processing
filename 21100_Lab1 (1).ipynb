```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

audio_signal, sampling_rate = librosa.load('Recording.wav')
plt.figure(figsize=(14, 5))
librosa.display.waveshow(audio_signal)
librosa.display.waveshow(audio_signal, color='black')
plt.title('Speech Signal Waveform')
plt.show()
```

```python
signal_length = len(audio_signal)

min_magnitude = np.min(audio_signal)
max_magnitude = np.max(audio_signal)

print("Length of the signal:", signal_length, "samples")
print("Sampling rate:", sampling_rate, "Hz")
print("Magnitude range:", min_magnitude, "to", max_magnitude)
```

```python
start_point = int(2 * sampling_rate)
end_point = int(3 * sampling_rate)

segment = audio_signal[start_point:end_point]

librosa.display.waveshow(segment, color='black')  
ipd.Audio(segment, rate=sampling_rate)
```

```python
segment_1_start_point = int(1 * sampling_rate)
segment_1_end_point = int(4 * sampling_rate)

segment_2_start_point = int(2 * sampling_rate)
segment_2_end_point = int(4 * sampling_rate)

segment_1 = audio_signal[segment_1_start_point:segment_1_end_point]
segment_2 = audio_signal[segment_2_start_point:segment_2_end_point]

print("Playing Segment 1:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1)
plt.title('Segment-1')
librosa.display.waveshow(segment_1, color='black') 
plt.show()
ipd.Audio(segment_1, rate=sampling_rate)
```

```python

print("Playing Segment 2:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2, color='black')  
plt.title('Segment-2')  
plt.show()

ipd.Audio(segment_2, rate=sampling_rate)
```
